```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv('laptop_price - dataset.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
variables_to_encode = [
    'Company','Product','TypeName','CPU_Company','GPU_Company','OpSys'
]

label_encoders = {}
for col in variables_to_encode:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le
```

```python
df.head()
```

```python
Columns =['ScreenResolution','Memory','GPU_Type','CPU_Type']


label_encoder = LabelEncoder()

for column in Columns:
    df[column] = label_encoder.fit_transform(df[column])

df.head()
```

```python
df.corr()
```

```python
correlations=df.corr()['Price (Euro)'].drop('Price (Euro)')

threshold=0.2
least_correlated_features=correlations[abs(correlations)<threshold].index

df_dropped=df.drop(columns=least_correlated_features)
df_dropped.head()
```

```python
print(list(least_correlated_features))
```

```python
X = df_dropped.drop(columns=['Price (Euro)']) 
y = df_dropped['Price (Euro)']
```

```python
scaler= StandardScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y.values.reshape(-1, 1)) 
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
lin_model=lin_reg.predict(X_test)
```

```python
mse = mean_squared_error(y_test,lin_model)
mae = mean_absolute_error(y_test,lin_model)
```
